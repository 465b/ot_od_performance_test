```python
def random_sample(n, x_min, x_max, y_min, y_max):
    x_samples = np.random.uniform(x_min, x_max, n)
    y_samples = np.random.uniform(y_min, y_max, n)
    return np.column_stack((x_samples, y_samples))
```

```python
# generate 100 points within the range
release_points_lon_lat = random_sample(
    100, 
    np.min(df.variables['lat']),np.max(df.variables['lat']),
    np.min(df.variables['lon']),np.max(df.variables['lon'])
    )
```
